```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_path = 'massage_2019-07-02_15-33-57.csv'
with open(data_path, 'r') as f:
    reader = csv.reader(f, delimiter=',')
    # get all the rows as a list
    data = list(reader)
    # transform data into numpy array
    data = np.array(data).astype(float)

print(data.shape)
display_begin = 200
display_limit = 600

# Plot the data
plt.figure(figsize = [20,10])
plt.plot(data[display_begin:display_limit, 0], data[display_begin:display_limit, 1])
plt.plot(data[display_begin:display_limit, 0], data[display_begin:display_limit, 2])


plt.gca().invert_yaxis()
plt.xlabel("Time")
plt.ylabel("Compression Depth")

plt.show()

#
#plt.xlabel('Table Index')
#plt.ylabel(headers[2])
#plt.show()
```

```python
plt.figure(figsize = [20,10])
plt.plot(data[display_begin:display_limit, 0], data[display_begin:display_limit, 2] - data[display_begin:display_limit, 1])

plt.xlabel("Time")
plt.ylabel("Difference between measures")

plt.show()
print(np.mean(data[display_begin:display_limit, 2] - data[display_begin:display_limit, 1]))
```
