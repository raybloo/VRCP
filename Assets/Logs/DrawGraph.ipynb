```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def find_index(data,lower,upper):
    low_ind = -1
    up_ind = -1
    for i, val in enumerate(data):
        if(val[0] >= lower and low_ind == -1):
            low_ind = i
        if(val[0] < upper):
            up_ind = i
    if(up_ind == -1):
        up_ind = len(data)
    if(low_ind == -1):
        low_ind = 0
    return (low_ind,up_ind)
        
```

```python
data_path = 'massage_log.csv'
with open(data_path, 'r') as f:
    reader = csv.reader(f, delimiter=',')
    # get all the rows as a list
    massage_data = list(reader)
    # transform data into numpy array
    massage_data = np.array(massage_data).astype(float)

print(massage_data.shape)

data_path = 'sensor_log.csv'
with open(data_path, 'r') as f:
    reader = csv.reader(f, delimiter=',')
    # get all the rows as a list
    sensor_data = list(reader)
    # transform data into numpy array
    sensor_data = np.array(sensor_data).astype(float)

print(sensor_data.shape)

massage_begin, massage_end = find_index(massage_data,150,170)
sensor_begin, sensor_end = find_index(sensor_data,150,170)
print(massage_end-massage_begin)
print(sensor_end-sensor_begin)
#print(sensor_begin)
#print(sensor_end)
# Plot the data
plt.figure(figsize = [20,10])
plt.plot(massage_data[massage_begin:massage_end, 0], massage_data[massage_begin:massage_end, 1])
plt.plot(massage_data[massage_begin:massage_end, 0], massage_data[massage_begin:massage_end, 2])
plt.plot(sensor_data[sensor_begin:sensor_end, 0], sensor_data[sensor_begin:sensor_end, 1])



plt.gca().invert_yaxis()
plt.xlabel("Time [s]")
plt.ylabel("Compression Depth [cm]")

plt.show()

#
#plt.xlabel('Table Index')
#plt.ylabel(headers[2])
#plt.show()
```
